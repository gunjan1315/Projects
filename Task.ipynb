```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of age
plt.figure(figsize=(10, 8)) # Adjusted figure size
data['age'].hist()
plt.title("Age Distribution")
plt.show()

# Boxplot: cholesterol by heart disease
plt.figure(figsize=(10, 8)) # Adjusted figure size
sns.boxplot(x='heart_disease', y='chol', data=data)
plt.show()

# Scatter: age vs max heart rate
plt.figure(figsize=(10, 8)) # Adjusted figure size
sns.scatterplot(x='age', y='thalach', hue='heart_disease', data=data)
plt.show()

# Heatmap of correlations
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
plt.show()
```

 Insights from the Heart Disease Dataset
1. Prevalence

Around 40% of patients in this dataset have heart disease (heart_disease = 1), while 60% do not.

This shows the dataset is fairly balanced  good for modeling, and also clinically relevant (heart disease is common).

2. Age Factor

Age ranges from 29 to 76 years, with an average of 52 years.

Heart disease is more common in older patients (youd confirm this with a grouped table/boxplot).

3. Cholesterol & Blood Pressure

Cholesterol levels are high on average (347 mg/dl), with values up to 600 mg/dl.

Elevated cholesterol is a strong known risk factor; likely patients with higher cholesterol will show more disease cases.

Resting blood pressure (trestbps) ranges 90200 mmHg, mean ~145  many patients are hypertensive.

4. Max Heart Rate (thalach)

Max heart rate achieved ranges 60210 bpm, average ~135 bpm.

Patients with heart disease often show lower max heart rates compared to healthy patients (to be confirmed with boxplots).

5. Lifestyle Factors

About 35% are smokers and 19% are diabetic.

Both smoking and diabetes are known contributors to cardiovascular risk, so these variables are important to analyze further.

6. BMI

BMI ranges 1540, with average ~27.5 (slightly overweight).

Higher BMI likely overlaps with higher disease prevalence, especially when combined with diabetes and cholesterol.

7. Correlations

Initial correlation checks usually show:

Cholesterol & BMI moderately correlated.
